#### 3. Create a data dictionary for the dataset


##### A data dictionary for a dataset is collection of descriptions of the data objects or items in a data model for the benefit of programmers and others who need to refer to them. In describing this dataset, the following structure can be used as a data dictionary: {[State]:'State Abbreviation', [Rate]:Rate (in float), [Verbal]:Verbal score (int), [Math]:Math score (int)}. 

##### Another way this dataset can be organized into a data dictionary is with each state as a key, and the values associated with each state, such as rate, verbal score, and math score, acting as the value in the key:value pair.

```python
# 4 and #5. load data into a list of list and print list

with open('sat_scores.csv', 'rU') as csvfile:
    reader = csv.reader(csvfile)
    sat_list = []
    for row in reader:
        sat_list.append(row)
print sat_list
```

```python
# 6. Extract a list of the labels from the data, and remove them from the data.
labels = sat_list[0]
del sat_list[0]
print labels
```

```python
# 7. Create a list of State names extracted from the data

state_name = []
for i in sat_list:
    state_name.append(i[0])
print state_name
```

```python
# 8. Print the types of each column

for item in sat_list:
    for i in item:
        print type(i)
    break
```

```python
# 9. Do any types need to be reassigned? If so, go ahead and do it.

print """
Yes, in order to analyze data, the current strings for 'Rate', 'Verbal', and 'Math' need to be converted into 
integers and/or floats.
"""

rate = []
for i in sat_list:
    rate.append(float(i[1]))
print rate
```

```python
verbal = []
for n in sat_list:
    verbal.append(int(n[2]))
print verbal
```

```python
math = []
for s in sat_list:
    math.append(int(s[3]))
print math
```

```python
# 10. Create a dictionary for each column mapping the State to its respective value for that column.

state_value = {}
for i in range(len(state_name)):
    state_value[state_name[i]] = [rate[i], verbal[i], math[i]]
print state_value
```

```python
# 13. compute standard deviation using line comprehension
import numpy as np

def stddev(lst):
    """returns the standard deviation of lst"""
    mean = sum(lst)/len(lst)
    variance = sum([(e-mean)**2 for e in lst])
    return np.sqrt(variance)

stddev(rate)
```

```python
def stddev(vscore):
    """returns the standard deviation of lst"""
    mean = sum(vscore)/len(vscore)
    variance = sum([(e-mean)**2 for e in vscore])
    return np.sqrt(variance)

stddev(verbal)
```

```python
def stddev(mscore):
    """returns the standard deviation of lst"""
    mean = sum(mscore)/len(mscore)
    variance = sum([(e-mean)**2 for e in mscore])
    return np.sqrt(variance)

stddev(math)
```

#### 17. What is the typical assumption for data distribution?

##### The typical assumption for data distribution is for the data to have a normal distribution. This means that if we were to take the average of a sample, record it, did another study and recorded that mean down, and repeat this procedure an infinite number of times, then the distribution of that mean would always be a perfect bell curve.

#### 18. Does that distribution hold true for our data?


##### According to the graphs above, it does not appear that we have a perfect bell curve for the data--hence we do not have a normal distribution of the data. If this is true, then we can perhaps assume that this data may not be representative of the entire population of SAT test-takers. The reasons for this can be varied.

##### Some states or school districts do not emphasize the SAT much as admissions test to college. If there were colleges that also do no emphasize or accept SAT scores for admissions, this may lower the rate in which students participate in the test as much as states that do emphasize the SAT. There may also be students who do not plan on going to college after high school, and therefore opt not to take the SATs. Another reason could also be that some students missed taking the exam on the day the test was administered and the data collected.
